## Aproximación por series de Taylor de la función seno y coseno

```python
from scipy.misc import derivative   # scipy: Cálculo científico y técnico
import math                         # paquete de matemáticas
```

```python
class TaylorSeries():
    def __init__(self, function, order, center=0):
        self.center = center
        self.f = function 
        self.order = order
        self.d_pts = order*2
        self.coefficients = []

        # number of points (order) for scipy.misc.derivative
        if self.d_pts % 2 == 0: # must be odd and greater than derivative order
            self.d_pts += 1

        self.__find_coefficients()

    def __find_coefficients(self):
        for i in range(0, self.order+1):
            self.coefficients.append(round(derivative(self.f, self.center, n=i, order=self.d_pts)/math.factorial(i), 5))
```

```python
def SeriesTaylor(f, ord, a = 0):
    """
    Función para calcular series de Taylor truncadas
    
    input:
    - f   : función a aproximar
    - ord : orden del truncamiento
    - a   : centro de la serie, por defecto centrado en a=0 (serie de Maclaurin)
    
    output:
    - coefs : coeficientes de la serie truncada. 
    """

    coefs = []
    for i in range(ord):
        coefs.append( round(derivative(f, a, n=i)/math.factorial(i), 5) )
    
    return coefs
```

```python
xx = np.linspace(-2*np.pi, 2*np.pi, 200)
ss = np.sin(xx)
cc = np.cos(xx)
```

```python
fig = plt.figure()
plt.subplots_adjust(wspace=1, hspace=0.5)

for i in range(2):
    ax = fig.add_subplot(2, 1, i+1)

    if i == 0:
        ax.plot(xx, ss, label="sin(x)")
        ax.set_title(r"$\sin x$")
    if i == 1:
        ax.plot(xx, cc, label="cos(x)")
        ax.set_title(r"$\cos x$")
    
    ax.grid(which='major', linestyle='-')
    ax.legend()

plt.show()
```
